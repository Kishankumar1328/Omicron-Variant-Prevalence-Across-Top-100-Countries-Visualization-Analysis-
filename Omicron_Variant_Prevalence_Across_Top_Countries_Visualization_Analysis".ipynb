```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/covid-cases-omicron.csv").head(100)
df
```

```python
sns.scatterplot(x="Entity",y="Omicron_percentage",data=df)
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
sns.barplot(x="Entity",y="Omicron_percentage",data=df)
plt.xticks(rotation=90)
```

```python
sns.lineplot(x="Entity",y="Omicron_percentage",data=df)
plt.xticks(rotation=90)
```

```python
sns.boxplot(x="Entity",y="Omicron_percentage",data=df)
plt.xticks(rotation=90)
```

```python
sns.violinplot(x="Entity",y="Omicron_percentage",data=df)
plt.xticks(rotation=90)
```

```python
pd.crosstab(df["Entity"],df["Omicron_percentage"]).plot(kind="area",stacked=False,figsize=(20,10))
plt.xlabel("countries affect by Omicron",fontsize=29)
plt.ylabel("Cases(percentage) ",fontsize=29)
plt.title("Omicron case",fontsize=42)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
```
